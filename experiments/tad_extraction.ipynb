```python
df = pd.read_csv(r"C:\Users\erick\Documents\Uni\Thesis\delay-replanning-slack\experiments\results\3000e-Hdr.csv", index_col=0)

    
def extract_tipping_point(df):
    def apply_func(df):
        df["tipping_point"]= (df["alpha"]>df["beta"].shift()).cumsum()
        result=df.groupby("tipping_point").agg({"alpha":"min", "beta": "max"})
        return result
    
    df["delay_location"] = df["delay_location"].str.split("|").apply(lambda x: x[0])
    df = df.groupby(by='delay_location').apply(apply_func, include_groups=False)
    df = df.loc[df["alpha"] < 900]
    return df

df.groupby(["scenario", "label"]).apply(extract_tipping_point, include_groups=False)
```
